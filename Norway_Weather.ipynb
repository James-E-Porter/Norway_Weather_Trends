# Title: Analyzing Yearly Averages of Meteorological Data in Norway


## Introduction:

This project analyzes the yearly averages of meteorological data from 55 meteorological stations from Norway from 2010 to 2021. Data from the Norwegian Meteorological Institute (https://www.met.no/) was parsed with the Frost API (https://frost.met.no/index.html). It includes the station id, coordinates, date, max and mean air temperature, max and mean wind speed, max and mean relative humidity, and sum of precipitation for each day.
This dataset - NorwayMeteoDataCompleted.csv was downloaded from Kaggle.com.

### Data Cleaning:

The data was preprocessed, so no data cleaning was needed. 

### Analysis:

In this section, the yearly averages of the meteorological data from the 55 meteorological stations in Norway will be analyzed. 

To begin, the Python library pandas was imported in order to read the data and manipulate it.

```python
import pandas as pd

#The data was then read into a pandas dataframe using the read_csv() function.

df = pd.read_csv('NorwayMeteoDataCompleted.csv')

#The data was then grouped by year, and the mean of all the columns was calculated for each year.

year_avg = df.groupby('year').mean()
```

### Visualizations:

To visualize the yearly averages, the Python library matplotlib was imported.

```python
import matplotlib.pyplot as plt
```

Then, a line plot was created to show the mean air temperature for each year.

```python
year_avg.plot(y='mean(air_temperature P1D)', kind='line', figsize=(7,3))

plt.title('Mean Air Temperature by Year')
plt.xlabel('Year')
plt.ylabel('Mean Air Temperature (°C)')
plt.show()

```

Next, a line plot was created to show the mean wind speed for each year.

```python
year_avg.plot(y='mean(wind_speed P1D)', kind='line', figsize=(7,3))

plt.title('Mean Wind Speed by Year')
plt.xlabel('Year')
plt.ylabel('Mean Wind Speed (m/s)')
plt.show()
```

Finally, a line plot was created to show the mean relative humidity for each year.

```python
year_avg.plot(y='mean(relative_humidity P1D)', kind='line',figsize=(7,3))

plt.title('Mean Relative Humidity by Year')
plt.xlabel('Year')
plt.ylabel('Mean Relative Humidity (%)')
plt.show()
```

### Conclusion:

The analysis of the yearly averages of meteorological data from Norway showed that the mean air temperature has stayed relatively steady over the past 10 years, however, , wind speed and relative humidity have had an upward trend.  Further analysis can be done to compare the data from different stations and to look at the data over a longer period of time.
